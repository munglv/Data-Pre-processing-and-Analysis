```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'sex':['male','female','female','male'], 'shape':['medium','small','small','large']})
df
```

### Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df['shape_'] = le.fit_transform(df['shape'])
df
```

```python
df['sex_'] = le.fit_transform(df['sex'])
df
```

```python
df = pd.DataFrame({'sex':['male','female','female','male'], 'shape':['medium','small','small','large']})
```

```python
cols = ['sex','shape']
```

```python
le = LabelEncoder()
for col in cols:
    df[col+'_'] = le.fit_transform(df[col])
```

```python
df
```

```python
df = pd.DataFrame({'sex':['male','female','female','male'], 'shape':['medium','small','small','large']})
```

### One-hot encoder

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
```

```python
ohe.fit(df[['sex']])
```

```python
df_sex_ohe = ohe.transform(df[['sex']]).toarray()
df_sex_ohe
```

```python
ohe.categories_
```

```python
df_ = pd.DataFrame(df_sex_ohe, columns=['S_'+i for i in ohe.categories_[0]])
df_
```

```python
dfOneHot = pd.concat([df,df_], axis=1)
dfOneHot
```
