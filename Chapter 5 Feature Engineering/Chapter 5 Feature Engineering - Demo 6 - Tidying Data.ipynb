```python
import numpy as np
import pandas as pd
```

```python
data = {"marker":["Delorian", "Fantom"],
       "cyl":[4, 2],
       "hp": [160, 80],
       "carb": [4,2]}
df = pd.DataFrame(data)
df
```

### unpivot

```python
# unpivot
df_new = pd.melt(frame=df, id_vars="marker", 
                 value_vars=["cyl", "hp", "carb"], var_name="metric")
df_new
```

```python
pd.melt(frame=df, id_vars="marker", var_name="metric")
```

### pivot

```python
# xoay dòng thành cột
df_pivot = df_new.pivot(index = 'marker', 
                        columns='metric', values='value')
df_pivot
```

```python
df_new = df_new.append({'marker' : 'Fantom' , 'metric' : 'carb', 'value' : 4} , ignore_index=True)
df_new
```

```python
df_new.pivot(index = 'marker', 
                        columns='metric', values='value')
```

```python
# Pivot table
df_pivot = df_new.pivot_table(index = 'marker', columns='metric', 
                                values='value', aggfunc= np.mean)
df_pivot
```

```python
# tạo dataframe df
df = pd.DataFrame({'name':['Tom','John','Mike'], 'Math':[7.8,8.5,5.0], 
                   'Chemist':[8.5,9.0,7.0], 'Physist':[8.9,9.5,6.0]})
df
```

```python
# yêu cầu 1: sử dụng pd.melt để chuyển df --> df_new
df_new = pd.melt(frame=df, id_vars='name', value_vars=['Math','Chemist','Physist'], var_name='subject')
df_new
```

```python
# yêu cầu 2: từ df_new tính điểm trung bình các môn
df_new.groupby('subject')['value'].mean()
```

```python
# yêu cầu 3: từ df_new tính điểm trung bình các học sinh
df_new.groupby('name')['value'].mean()
```

```python
# yêu cầu 4: từ df tính điểm trung bình các môn
df.mean(axis=0)
```

```python
# yêu cầu 5: từ df tính điểm trung bình các học sinh
df.mean(axis=1)
```
