```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
exam = pd.read_csv("data/exam.csv")
```

```python
X = exam[['hours']]
y = exam['pass']
```

```python
from sklearn.model_selection import train_test_split   
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
print('Train Accuracy Score:', model.score(X_train, y_train))
print('Test Accuracy Score:', model.score(X_test, y_test))
```

```python
from sklearn.metrics import accuracy_score 
```

```python
yhat_train = model.predict(X_train)
print("Train Accuracy:", accuracy_score(y_train, yhat_train)) 
yhat_test = model.predict(X_test)
print("Test Accuracy:", accuracy_score(y_test, yhat_test)) 
```

## Thực hiện cross-validation

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
exam = pd.read_csv("data/exam.csv")
```

```python
X = exam[['hours']]
y = exam['pass']
```

```python
# Khởi tạo mô hình Logistic Regression
model = LogisticRegression()

# Thực hiện 5-fold cross-validation
scores = cross_val_score(model, X, y, cv=5, scoring='accuracy')
```

```python
# In kết quả cross-validation
print("Accuracy scores for each fold:", scores)
print("Mean accuracy:", np.mean(scores))
```
