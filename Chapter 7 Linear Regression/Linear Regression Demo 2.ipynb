### Polynomial Regression
- Nếu các điểm dữ liệu của bạn không phù hợp với hồi quy tuyến tính (một đường thẳng đi qua tất cả các điểm dữ liệu), thì đó có thể là lý tưởng cho hồi quy đa thức.
- Hồi quy đa thức, giống như hồi quy tuyến tính, sử dụng mối quan hệ giữa các biến x và y để tìm ra cách tốt nhất để vẽ một đường cong đi qua các điểm dữ liệu.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.metrics import r2_score
```

```python
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]
```

```python
np.corrcoef(x,y) # tương quan thuận và yếu
```

```python
plt.scatter(x, y)
plt.show()
```

```python
mymodel = np.poly1d(np.polyfit(x, y, 2)) # đường cong bậc 2

plt.scatter(x, y)
plt.plot(x, mymodel(x))
plt.show()
```

```python
mymodel = np.poly1d(np.polyfit(x, y, 3)) # đường cong bậc 3

plt.scatter(x, y)
plt.plot(x, mymodel(x))
plt.show()
```

```python
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]
```

```python
mymodel = np.poly1d(np.polyfit(x, y, 2))
print(r2_score(y, mymodel(x)))
```

```python
mymodel = np.poly1d(np.polyfit(x, y, 3))
print(r2_score(y, mymodel(x)))
```
