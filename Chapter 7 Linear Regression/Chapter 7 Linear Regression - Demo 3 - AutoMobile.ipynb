```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/automobileEDA.csv')
```

```python
X = df[['highway-mpg']]
y = df['price']
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pr = PolynomialFeatures(degree=2) # Khởi tạo PolynomialFeatures object với degree (ví dụ = 2)
```

```python
X_pr = pr.fit_transform(X)
```

```python
X.shape, X_pr.shape # Dữ liệu gốc có 201 mẫu, 1 thuộc tính => dữ liệu transform có 201 mẫu, 3 thuộc tính 
```

```python
X.head()
```

```python
X_pr[:5] # y = aX^0 + b1X^1 + b2X^2 # X^0 = 1
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pr, y, random_state=42, test_size=0.2)
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
print('The full R-square is:', lm.score(X_pr,y))
print('The train R-square is:', lm.score(X_train,y_train))
print('The test R-square is:', lm.score(X_test,y_test))
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
yhat = lm.predict(X_test)
```

```python
mse = mean_squared_error(y_test, yhat)
mae = mean_absolute_error(y_test, yhat)
print('The MSE of price and predicted value is: ', mse)
print('The MAE of price and predicted value is: ', mae)
```
