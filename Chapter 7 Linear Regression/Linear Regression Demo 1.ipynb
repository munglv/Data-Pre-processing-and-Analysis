### Linear Regression
- Hồi quy tuyến tính sử dụng mối quan hệ giữa các điểm dữ liệu để vẽ một đường thẳng đi qua tất cả chúng.
- Đường này có thể được sử dụng để dự đoán các giá trị trong tương lai.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.metrics import r2_score
```

## Tập dữ liệu có hồi quy tuyến tính

```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6] # số năm mua xe
y = [99,86,87,88,111,86,103,87,94,78,77,85,86] # vận tốc trung bình của xe
```

```python
np.corrcoef(x,y) # tương quan khá mạnh và nghịch
```

```python
plt.scatter(x, y)
plt.show()
```

```python
slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))
```

```python
plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

```python
print('slope:', slope)
print('intercept:', intercept)
print('r^2:', r*r)      
```

```python
print(r2_score(y, mymodel))
```

## Tập dữ liệu không có hồi quy tuyến tính

```python
x = [89,43,36,36,95,10,66,34,38,20,26,29,48,64,6,5,36,66,72,40]
y = [21,46,3,35,67,95,53,72,58,10,26,34,90,33,38,20,56,2,47,15]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

```python
print('r^2:', r*r)
```

```python
print(r2_score(y, mymodel))
```
