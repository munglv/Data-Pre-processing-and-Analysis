```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('data/automobileEDA.csv')
```

```python
X = df[['horsepower','curb-weight','engine-size','highway-mpg']]
y = df['price']
```

```python
df[['horsepower','curb-weight','engine-size','highway-mpg','price']].corr()
```

```python
sns.regplot(data=df, x='horsepower', y='price');
```

```python
sns.regplot(data=df, x='engine-size', y='price');
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pr = PolynomialFeatures(degree=2)
```

```python
X_pr = pr.fit_transform(X)
```

```python
X.shape, X_pr.shape
```

```python
# X1, X2, X3, X4 ==>
# X1, X2, X3, X4, X1^2, X2^2, X3^2, X4^2
# X1X2, X1X3, X1X4, X2X3, X2X4, X3X4
```

```python
X_pr[:2]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pr, y, random_state=42, test_size=0.2)
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
yhat = lm.predict(X_test)
```

```python
print('The full R-square is:', lm.score(X_pr,y))
print('The train R-square is:', lm.score(X_train,y_train))
print('The test R-square is:', lm.score(X_test,y_test))
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
mse = mean_squared_error(y_test, yhat)
mae = mean_absolute_error(y_test, yhat)
print('The MSE of price and predicted value is: ', mse)
print('The MAE of price and predicted value is: ', mae)
```
