## Kiểm định Chi-square dùng cho 2 biến phân loại
### Phát biểu giả thuyết
- H0: 2 biến phân loại là độc lập
- Ha: 2 biến phân loại là phụ thuộc

### Pclass có ảnh hưởng đến Survived hay là không?

```python
import pandas as pd
from scipy.stats import chi2_contingency
```

```python
# Đọc dữ liệu từ file CSV
titanic_data = pd.read_csv('data/titanic.csv') # mẫu dùng cho kiểm định
```

```python
# Tạo bảng tần số cho Pclass và Survived
cross_tab = pd.crosstab(titanic_data['Pclass'], titanic_data['Survived'])
```

```python
cross_tab
```

```python
# Phát biểu giả thuyết:
# H0: Pclass và Survived là hai biến độc lập
# Ha: Pclass và Survived là hai biến phụ thuộc
```

```python
# Thực hiện kiểm định Chi-square
chi2, p, dof, expected = chi2_contingency(cross_tab)
```

```python
# In kết quả
print("Giá trị Chi-square:", chi2)
print("Giá trị p-value:", p)
print("Bậc tự do:", dof)
print("Bảng tần số dự kiến:")
print(expected)
```

```python
# nếu p-value<0.05 --> bác bỏ H0 và chấp nhận Ha
# nếu p-value>=0.05 --> chấp nhận H0
# kết luận: do p-value<0.05 ==> Pclass và Survived là hai biến phụ thuộc và có ý nghĩa thống kê (không phải do ngẫu nhiên)
# khoảng tin cậy 95%: nghĩa là trong 100 lần lấy mẫu thì xem như kết luận là đúng với 95 lần
```
