```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
sns.set()
```

# Tập dữ liệu Titanic
## Tìm outlier theo công thức Quantiles
- Đọc dữ liệu titanic.csv
- Vẽ boxplot cho biến số Age để xem có outlier?
- Tìm ra các outlier và xử lý: sao chép, xóa 

```python
titanic = pd.read_csv('data/titanic.csv')
```

```python
titanic.head()
```

```python
titanic.Age.describe()
```

```python
sns.boxplot(data=titanic, y='Age'); # có outlier
```

```python
sns.boxplot(data=titanic, y='Age', whis=1.8); 
```

```python
titanic.Age.skew() # >0 lệch phải
```

```python
m = titanic.Age.mean()
sns.distplot(titanic.Age) # displot
plt.axvline(x=m, color='r')
plt.show()
# lệch phải do có đóng góp của outlier (mean > mode, median)
```

# Phát hiện các outlier của biến Age và xử lý

```python
# xác định các outlier
Q1 = titanic.Age.quantile(0.25)
Q2 = titanic.Age.quantile(0.50)
Q3 = titanic.Age.quantile(0.75)
IQR = Q3 - Q1
print('Q1 25%:', round(Q1,2))
print('Q2 50% (median):', round(Q2,2))
print('Q3 75%:', round(Q3,2))
print('IQR:', round(IQR,2))
```

```python
# có ít nhất 25% hành khách có tuổi <= 20
# có ít nhất 75% hành khách có tuổi >= 20
```

```python
w = 1.5
UW = Q3 + w*IQR
print('upper whisker:', round(UW,2))
```

```python
# xác định các upper outlier
outliers = titanic[titanic['Age']>UW].sort_values(by='Age')
outliers['Age']
```

```python
# xóa các outlier
titanic.drop(index=outliers.index, inplace=True)
```

```python
sns.boxplot(data=titanic, y='Age'); # không còn outlier
```

# Tập dữ liệu Insurance
## Tìm outlier theo ngoài khoảng độ lệch chuẩn
- Đọc dữ liệu insurance.csv
- Vẽ boxplot cho biến số bmi để xem có outlier?
- Tìm ra các outlier và xử lý: sao chép, xóa 

```python
insurance = pd.read_csv('data/insurance.csv')
```

```python
insurance.head()
```

```python
insurance.shape
```

```python
sns.boxplot(data=insurance, y='bmi'); # có outliers
```

```python
insurance.bmi.describe()
```

```python
m = insurance.bmi.mean()
sns.distplot(insurance['bmi']); # xấp xĩ phân phối chuẩn
plt.axvline(x=m, color='r')
plt.show()
```

```python
from scipy.stats import zscore

insurance['bmi_zscore'] = insurance[['bmi']].apply(zscore) # chuyển đổi zscore --> phân phối chuẩn tắc (Standard Normal Distribution)
# phân phối chuẩn tắc có trung bình là 0, độ lệch chuẩn là 1
```

```python
sns.distplot(insurance['bmi_zscore']); # các outlier nằm về phía bên phải (ngoài 3 độ lệch chuẩn)
```

```python
# xác định các upper outlier
r = 3.0
outliers = insurance[(insurance['bmi_zscore']<-r)|(insurance['bmi_zscore']>r)]
outliers['bmi_zscore'].sort_values()
```

```python
outliers
```

```python
insurance.drop(index=outliers.index, inplace=True) # xóa các outlier
```

```python
sns.boxplot(data=insurance, y='bmi');
```

```python
sns.boxplot(data=insurance, y='bmi', whis=1.7);
```
