### Tập dữ liệu IRIS

```python
import glob
```

```python
csv_files = glob.glob('data/iris*.csv') # 'data/*.csv'
```

```python
csv_files
```

```python
lst = []
for f in csv_files:
    df_ = pd.read_csv(f, index_col=0)
    lst.append(df_)
```

```python
iris = pd.concat(lst, ignore_index=True)
```

```python
iris.shape
```

```python
iris.head()
```
