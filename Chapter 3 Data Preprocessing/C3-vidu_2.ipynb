```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
sns.set()
```

```python
titanic = pd.read_csv('data/titanic.csv')
```

```python
titanic.head()
```

```python
# biến output: biến Survived
# biến input: là các biến còn lại
```

```python
# các biến có ý nghĩa phân tích gồm:
cols = ['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked','Survived']
titanic = titanic[cols]
titanic.head()
```

```python
# các biến object
cat_cols = titanic.select_dtypes(include='object').columns
cat_cols
```

```python
for col in cat_cols:
    print(col,'-',titanic[col].unique())
```

```python
# biến phân loại là: Sex, Embarked
```

```python
# các biến number
num_cols = titanic.select_dtypes(include='number').columns
num_cols
```

```python
for col in num_cols:
    if titanic[col].nunique()<=10:
        print(col, '-',titanic[col].unique())
    else:    
        print(col, '-',titanic[col].nunique())
# biến phân loại là: Pclass, Survived        
```
