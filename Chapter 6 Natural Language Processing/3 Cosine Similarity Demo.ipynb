Độ tương tự cosine là phương pháp thường được sử dụng để đo độ tương tự giữa hai vectơ trong không gian nhiều chiều. Trong ngữ cảnh Xử lý ngôn ngữ tự nhiên (NLP), nó thường được sử dụng để so sánh sự giống nhau giữa các tài liệu văn bản hoặc các câu được biểu diễn dưới dạng vectơ. Nó tính toán cosin của góc giữa hai vectơ, trong đó góc nhỏ hơn biểu thị độ tương tự cao hơn. Độ tương tự cosine dao động từ -1 đến 1, với 0 biểu thị không có sự tương đồng.<br/>
Công thức này hoạt động bằng cách đo góc giữa hai vector A và B trong không gian nhiều chiều.Nếu góc giữa chúng gần với 0 độ, tức là chúng rất giống nhau, thì Cosine Similarity gần bằng 1. Nếu góc giữa chúng gần với 90 độ, tức là chúng rất khác nhau, thì Cosine Similarity gần bằng 0. Cosine Similarity có giá trị trong khoảng [-1, 1], trong đó:<br/>
1: Hai vector hoàn toàn tương đồng.
0: Hai vector không có tương đồng.
-1: Hai vector hoàn toàn ngược nhau.

```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Sample text data
documents = [
    "I love machine learning",
    "Machine learning is fascinating",
    "NLP is a subfield of AI",
    "Python is a popular programming language",
]
```

```python
# Step 1: Vectorize the text data using CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(documents)
```

```python
# Step 2: Calculate the cosine similarity between vectors
cosine_similarities = cosine_similarity(X, X)
```

```python
# Print the cosine similarity matrix
print(cosine_similarities)
```

```python
# Document to compare
target_document = "I am interested in machine learning"

# Step 1: Vectorize the text data using CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(documents)

# Vectorize the target document
target_vector = vectorizer.transform([target_document])

# Step 2: Calculate the cosine similarity between the target document and all other documents
cosine_similarities = cosine_similarity(target_vector, X)

# Print the cosine similarity scores
print("Cosine Similarity Scores:")
for i, score in enumerate(cosine_similarities[0]):
    print(f"Target Document vs. Document {i + 1}: {score}")
```

```python
# Flatten the cosine similarity scores and get the indices that sort them in descending order
similarity_scores = cosine_similarities[0]
sorted_indices = np.argsort(similarity_scores)[::-1]

# Get the top N most similar documents
top_N = 2  
top_documents = [documents[i] for i in sorted_indices[:top_N]]
top_scores = [similarity_scores[i] for i in sorted_indices[:top_N]]

# Print the top similar documents and their scores
print(f"Top {top_N} Similar Documents:")
for i, (doc, score) in enumerate(zip(top_documents, top_scores)):
    print(f"Document {i + 1}: {doc} (Similarity Score: {score})")
```
